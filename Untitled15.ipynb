```python
# Импортируем библиотеку pandas для работы с таблицами (DataFrame)
import pandas as pd

# Чтение CSV-файла с данными по пролонгациям
# df — DataFrame, содержащий информацию о коэффициентах пролонгации по месяцам и менеджерам
df = pd.read_csv('prolongations.csv')
df.head()
```

```python
# Чтение CSV-файла с финансовыми данными
# # df2 — DataFrame, содержащий финансовую информацию по менеджерам, возможно продажи, выручка и т.д.
df2 = pd.read_csv('financial_data.csv')
df2.head()
```

```python
# Объединяем два DataFrame по общему столбцу 'id'
merge_df = pd.merge(df, df2, on='id')
merge_df.head()
```

```python
# Создаем новый столбец 'Manager'
# merge_df['AM'] — основной столбец с именем менеджера
# merge_df['Account'] — запасной столбец, если AM пустой (NaN)
# combine_first выбирает значение из 'AM', если оно есть, иначе берёт из 'Account'
merge_df['Manager'] = merge_df['AM'].combine_first(merge_df['Account'])
merge_df.head()
```

```python
# Удаляем ненужные столбцы 'AM' и 'Account'
# # axis=1 указывает, что удаляем столбцы
merge_df = merge_df.drop(['AM', 'Account'], axis=1)
merge_df.head()
```

```python
# Приводим значения в столбце 'month' к правильному формату заглавных букв
merge_df['month'] = merge_df['month'].str.title()

# Создаём список всех месяцев, которые есть в данных
all_months = list(merge_df.columns[3:-1])

# Проверка наличия всех ожидаемых месяцев в DataFrame
# Если столбец с каким-то месяцем отсутствует, выводим предупреждение
for month in all_months:
    if month not in merge_df.columns:
        print(f"Столбец '{month}' отсутствует!")

# Функция для проверки, был ли проект остановлен ('стоп' или 'end') до или в последнем месяце реализации
def is_project_stopped(row):
    last_month = row['month']  # Определяем последний месяц реализации проекта/идентификатора
    last_month_index = all_months.index(last_month)  # Находим индекс этого месяца в списке всех месяцев
    months_to_check = all_months[:last_month_index+1]  # Берём все месяцы до и включая последний месяц реализации
    # Проверяем, встречается ли в этих месяцах отметка 'стоп' или 'end'
    for month in months_to_check:
        if row[month] in ['стоп', 'end']:
            return True  # Если проект остановлен, возвращаем True
    return False  # Иначе проект активен

# Применяем функцию ко всем строкам merge_df и исключаем строки с остановленными проектами
# ~ означает отрицание, т.е. выбираем только те строки, где проект не остановлен
df_cleaned = merge_df[~merge_df.apply(is_project_stopped, axis=1)]

df_cleaned.head(20)
```

```python
# Смотрим тип данных
df_cleaned.info()
```

```python
# Функция для очистки и преобразования значений в числовой формат
def clean_column(col):
    # Если значение в ячейке текстовое
    if isinstance(col, str):
        # Убираем лишние символы: неразрывные пробелы (\xa0), обычные пробелы, заменяем запятую на точку
        col = col.replace('\xa0', '').replace(' ', '').replace(',', '.')
        # Специальные текстовые значения
        if col.lower() == 'вноль':  # если написано 'вноль', присваиваем 0.0
            return 0.0
        elif col in ['стоп', 'end']:  # если проект остановлен, возвращаем None (чтобы пропуск учесть как NaN)
            return None
        else:
            return float(col)  # преобразуем текстовое число в float
    return col  # если значение уже число, оставляем без изменений

# Выбираем диапазон столбцов с месяцами (с 4-го до предпоследнего)
months = df_cleaned.columns[3:-1]

# Применяем функцию очистки ко всем ячейкам по каждому месяцу
for i in months:
    df_cleaned[i] = df_cleaned[i].apply(clean_column)

# Проверяем информацию о DataFrame после преобразования: типы данных, количество непустых значений
df_cleaned.info()

df_cleaned.head(20)
```

```python
# Убираем Причину дубля
df_agg = df_cleaned.drop(columns=['Причина дубля'])
df_agg.head()
```

```python
# Группируем данные по уникальному идентификатору проекта, месяцу и имени менеджера
# as_index=False — сохраняем группы как обычные столбцы, а не делаем их индексами
# sum(numeric_only=True) — суммируем только числовые столбцы для каждой группы
df_agg = df_agg.groupby(['id', 'month', 'Manager'], as_index=False).sum(numeric_only=True)
```

```python
# Список месяцев и словарь с индексами для сравнения
months = list(df_agg.columns[3:])
month_ind = {month: i for i, month in enumerate(months)}

# Преобразуем таблицу в длинный формат
df_long = df_agg.melt(id_vars=['id', 'month', 'Manager'], value_vars=months,
                      var_name='Месяц', value_name='Сумма')

# Фильтруем ненулевые и не NaN значения
df_long = df_long[df_long['Сумма'].notna() & (df_long['Сумма'] != 0)]

# Добавляем индексы месяцев для удобства
df_long['index_месяц'] = df_long['Месяц'].map(month_ind)
df_long['index_окончания'] = df_long['month'].map(month_ind)

# Фильтруем строки по условию:
# месяц отгрузки должен быть в пределах месяца окончания и два следующих месяца
df_long['valid'] = df_long.apply(
    lambda row: row['index_окончания'] <= row['index_месяц'] <= row['index_окончания'] + 2,
    axis=1
)

# Оставляем только те строки, где valid == True
df_long = df_long[df_long['valid']]

# Убираем временную колонку 'valid'
df_long = df_long.drop(columns=['valid'])

# Сортируем так, чтобы при drop_duplicates оставить менеджера с минимальным месяцем окончания (т.е. ближайшим)
df_long = df_long.sort_values(by=['id', 'Месяц', 'index_окончания'], ascending=[True, True, True])

# Оставляем по одному менеджеру на (id, Месяц)
df_clean = df_long.drop_duplicates(subset=['id', 'Месяц'], keep='first')

# Выбираем и переименовываем колонки
df_clean = df_clean[['id', 'Месяц', 'Manager', 'month', 'Сумма']].rename(columns={
    'Manager': 'Менеджер',
    'month': 'Месяц_окончания'
})

# Сортируем для удобного просмотра
df_clean = df_clean.sort_values(by=['Месяц'])

# Проверим для id == 31
print(df_clean[df_clean['Менеджер'] == 'Федорова Марина Васильевна'])
df_clean.head(1000)
```

```python
def find_coef_with_year(df_clean):
    result = []
    months = list(df_agg.columns[3:])  # Месяцы отгрузки

    # Проходим по каждому менеджеру
    for manager in df_clean['Менеджер'].unique():
        df_manager = df_clean[df_clean['Менеджер'] == manager]

        # Итоговые значения для годовых коэффициентов
        weighted_coef_1_sum, weighted_coef_2_sum, total_shipment_sum = 0, 0, 0

        # Проходим по каждому месяцу окончания
        for month_end in df_manager['Месяц_окончания'].unique():
            # Проекты, завершившиеся в month_end
            projects_end_month = df_manager[df_manager['Месяц_окончания'] == month_end]
            total_end_month = projects_end_month['Сумма'].sum()  # Сумма всех отгрузок в завершённом месяце

            # Индекс месяца окончания
            if month_end not in months:
                print(f"Месяц окончания '{month_end}' не найден в списке месяцев отгрузки")
                continue

            month_index = months.index(month_end)
            next_month = months[month_index + 1] if month_index + 1 < len(months) else None
            second_month = months[month_index + 2] if month_index + 2 < len(months) else None

            # Коэффициент 1: Пролонгации в следующий месяц
            if next_month:
                extended_projects = projects_end_month[projects_end_month['Месяц'] == next_month]
                extended_shipment = extended_projects['Сумма'].sum()
            else:
                extended_shipment = 0

            coef_1 = extended_shipment / total_end_month if total_end_month > 0 else 0

            # Коэффициент 2: Пролонгации во второй месяц
            if second_month:
                non_extended_projects = projects_end_month[
                    (projects_end_month['Месяц'] != next_month) &
                    (projects_end_month['Месяц_окончания'] == month_end)
                ]
                second_extended_projects = non_extended_projects[non_extended_projects['Месяц'] == second_month]
                second_extended_shipment = second_extended_projects['Сумма'].sum()
            else:
                second_extended_shipment = 0

            coef_2 = second_extended_shipment / total_end_month if total_end_month > 0 else 0

            # Обновляем итоговые суммы для расчёта средневзвешенных коэффициентов
            weighted_coef_1_sum += coef_1 * total_end_month
            weighted_coef_2_sum += coef_2 * total_end_month
            total_shipment_sum += total_end_month

            # Добавляем строку для месяца
            result.append({
                'Менеджер': manager,
                'Месяц_окончания': month_end,
                'Коэффициент_1': coef_1,
                'Коэффициент_2': coef_2
            })

        # После прохода по всем месяцам для данного менеджера вычисляем годовой коэффициент
        if total_shipment_sum > 0:
            coef_1_year = weighted_coef_1_sum / total_shipment_sum
            coef_2_year = weighted_coef_2_sum / total_shipment_sum
        else:
            coef_1_year, coef_2_year = 0, 0

        # Добавляем строку с годовыми коэффициентами для менеджера
        result.append({
            'Менеджер': manager,
            'Месяц_окончания': 'Итого за год',
            'Коэффициент_1': coef_1_year,
            'Коэффициент_2': coef_2_year
        })

    # Создаём итоговый DataFrame
    return pd.DataFrame(result)

# Пример использования функции
coef_df = find_coef_with_year(df_clean)

# Печать первых 30 строк
print(coef_df.head(30))

# Печать коэффициентов для конкретного менеджера
print(coef_df[coef_df['Менеджер'] == 'Федорова Марина Васильевна'])
```

```python
def find_coef_department(df_clean):
    result = []

    # Получаем список месяцев отгрузки
    months = list(df_agg.columns[3:])

    # Итоговые коэффициенты для всего отдела
    weighted_coef_1_sum, weighted_coef_2_sum, total_shipment_sum = 0, 0, 0

    # Проходим по каждому месяцу окончания
    for month_end in df_clean['Месяц_окончания'].unique():
        # Фильтруем все проекты, завершившиеся в month_end
        projects_end_month = df_clean[df_clean['Месяц_окончания'] == month_end]
        total_end_month = projects_end_month['Сумма'].sum()  # Сумма отгрузок в завершённом месяце

        # Проверка на наличие месяца окончания в списке месяцев отгрузки
        if month_end not in months:
            print(f"Месяц окончания '{month_end}' не найден в списке месяцев отгрузки")
            continue

        month_index = months.index(month_end)
        next_month = months[month_index + 1] if month_index + 1 < len(months) else None
        second_month = months[month_index + 2] if month_index + 2 < len(months) else None

        # Коэффициент 1: Пролонгации в следующий месяц
        if next_month:
            extended_projects = projects_end_month[projects_end_month['Месяц'] == next_month]
            extended_shipment = extended_projects['Сумма'].sum()
        else:
            extended_shipment = 0

        coef_1 = extended_shipment / total_end_month if total_end_month > 0 else 0

        # Коэффициент 2: Пролонгации во второй месяц
        if second_month:
            non_extended_projects = projects_end_month[
                (projects_end_month['Месяц'] != next_month) &
                (projects_end_month['Месяц_окончания'] == month_end)
            ]
            second_extended_projects = non_extended_projects[non_extended_projects['Месяц'] == second_month]
            second_extended_shipment = second_extended_projects['Сумма'].sum()
        else:
            second_extended_shipment = 0

        coef_2 = second_extended_shipment / total_end_month if total_end_month > 0 else 0

        # Добавляем данные для каждого месяца
        result.append({
            'Месяц_окончания': month_end,
            'Коэффициент_1_месяц': coef_1,
            'Коэффициент_2_месяц': coef_2
        })

        # Обновляем суммы для расчёта средневзвешенного коэффициента
        weighted_coef_1_sum += coef_1 * total_end_month
        weighted_coef_2_sum += coef_2 * total_end_month
        total_shipment_sum += total_end_month

    # Рассчитываем средневзвешенные коэффициенты для всего отдела за год
    if total_shipment_sum > 0:
        coef_1_year = weighted_coef_1_sum / total_shipment_sum
        coef_2_year = weighted_coef_2_sum / total_shipment_sum
    else:
        coef_1_year, coef_2_year = 0, 0

    # Добавляем строку с итоговыми коэффициентами для всего отдела за год
    result.append({
        'Месяц_окончания': 'Итого за год',
        'Коэффициент_1_месяц': coef_1_year,
        'Коэффициент_2_месяц': coef_2_year
    })

    return pd.DataFrame(result)

# Пример использования функции
department_coef_df = find_coef_department(df_clean)
print(department_coef_df)
```
