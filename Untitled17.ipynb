```python
import numpy as np

def initialisation(m, n):

  A = np.random.randn(m, n+1)
  B = np.ones((m,1))
  A_B = np.concatenate((A, B), axis = 1)
  return A_B

M = initialisation(3, 4)
print(M)
```

```python
import matplotlib.pyplot as plt
import numpy as np

dataset ={f"experience{i}": np.random.randn(100) for i in range(4)}

def graphique(data):
  n = len(data)
  plt.figure(figsize = (12, 8))

  for k, i in zip(data.keys(), range(1, n+1)):
      plt.subplot(n, 1, i)
      plt.plot(data[k])
      plt.title(k)

  plt.show()
```

```python
graphique(dataset)
```
